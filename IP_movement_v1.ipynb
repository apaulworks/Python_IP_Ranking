### Duplication of patents amongst parent company and subsidiaries
#### Detection, Display, and Removal

```python
#imports data from subsidiaries csv file.
import csv
import pandas as pd
parent_company = 'pfizer'      #this should match the name of .csv for the patents
file_str = "C:/Users/APL29/Documents/CODE/Patent_pulls/Pharma_subsidiaries/" + parent_company +".csv";
df = pd.DataFrame()
cols = []

#create a DataFrame from the read csv data.
df = pd.read_csv(file_str);
cols = df.columns.tolist()
df = df.drop(columns=cols[0])
df
```

```python
#increment through DataFrame entries to find matching Patent Numbers
col_name = 'patNum'  #'applNum'
j = 0; p = 0;
cnt = []; idx = []; idx_a = [];

for apn in df[col_name]:
    for rw in df[col_name]:
        if apn == rw:  
            idx_a.append(p)
            j += 1 
        p += 1;
    cnt.append(j)
    idx.append(idx_a)
    j = 0; p = 0; idx_a = [];

#organize matching patents by appending match count and match indicies. 
df[col_name + '_dupls'] = cnt;
df[col_name + '_index'] = idx;
```

```python
df
```

```python
#specifically for patents, I want to see what causes duplicates in this dataframe. I know that duplicates
#arise from having scraped both Pharma patent assignees and assignors.

dups = []; #initialize the array to capture duplicate indicies
#start by gathering duplicates
for idxa in df['patNum_index']:
    if len(idxa) > 1:            #if there exists duplicates
        dups.append(idxa)

#test assumption that patents with the same number are actually the same. Test a few

cc = 0; covered = [];
df2 = pd.DataFrame(); df3 = pd.DataFrame();
for nds in dups:
    for un in nds:
        if un not in covered:
            df2 = df2.append(df.iloc[[un]]);
            covered.append(un)
            if df.iloc[un]['inventionTitle'] != df.iloc[nds[0]]['inventionTitle']: # check here if invention name differs
                print('!! ', un , ' not ', nds[0])

    df3 = df3.append(df2.sort_values('patAssignorExDate'))    #sorting by Execution date of transfer, if applicable.
    df2 = df2.iloc[0:0]      #clear tmp dataframe
    cc += 1;
    
df3
```
